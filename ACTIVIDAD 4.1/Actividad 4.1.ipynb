```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('Ventas_totales.csv') 
```

```python
#Verificamos información del DataFrame
data.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Creo 2 dataframe para poder procesar los outliers
cuantitativas= data1.iloc[ : , 1:23 ]
cualitativas= data1.iloc[ : , [0] ]
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
#MÉTODO CUARTÍLICO: CONSIDERA MÁS ATÍPICOS
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos

#ESTE MÉTODO CONSIDERA MENOS ATÍPICOS


y=cuantitativas
Limite_Superior= y.mean() + 3*y.std() #SE AGREGAN 3 DESVIACIONES ESTANDAR 
Limite_Inferior= y.mean() - 3*y.std() #SE RESTAN 3 STV PARA ASEGURAR EL 95.5% DE LA INFO.
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
#CONDICIÓN PARA ELIMINAR LOS OUTLIERS FUERA DE LOS LÍMITES
data3= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
Datos_limpios_stdv = pd.concat([cualitativas, data_clean], axis=1)
Datos_limpios_stdv
```

```python
#Convertir DataFrame a csv
Datos_limpios_stdv.to_csv('Datos_limpios_stdv.csv')
```

EL MÉTODO DE DESVIACIONES ESTANDAR SE APLICA SOLO UNA VEZ, PORQUE RECALCULA LOS LÍMITES INTERCUARTÍLICOS, Y SIEMPRE SEGUIRÁN SALIENDO NUEVOS OUTLIERS POR LAS MEDIDAS NUEVAS. NO SE PUEDE COMPROBAR CON CAJA DE BIGOTES.

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**


```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
# Unimos el dataframe cuantitativo limpio con el dataframe cualitativo

Datos_limpios_iqr = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios_iqr.isnull().sum()
valores_nulos
```
